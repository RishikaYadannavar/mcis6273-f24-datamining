```python
import pandas as pd
data_url = 'https://ourairports.com/countries/US/airports.hxl'
# Load the CSV file into a DataFrame
airport_df = pd.read_csv(data_url)

# Remove the non-useful row (metadata row) if it exists
if airport_df.iloc[0, 0] == "#meta +id":
    airport_df = airport_df.drop(0)

# Save the cleaned DataFrame to a new CSV file named 'all_us_airports.csv'
airport_df.to_csv('all_us_airports.csv', index=False)
```

```python
import folium

# Load the filtered DataFrame
large_airports = pd.read_csv('large_us_airports.csv')

# Create a map centered at an initial location
map_center = [large_airports['latitude_deg'].mean(), large_airports['longitude_deg'].mean()]
mymap = folium.Map(location=map_center, zoom_start=5)

# Add markers for each large airport
for index, row in large_airports.iterrows():
    # Define popup content with airport name and IATA code
    popup_content = f"{row['name']} ({row['iata_code']})"

    # Add marker with custom icon and popup
    folium.Marker(location=[row['latitude_deg'], row['longitude_deg']],
                  popup=popup_content,
                  icon=folium.Icon(icon='plane', prefix='fa')).add_to(mymap)

# Save the map as an HTML file
mymap.save('large_airports_map.html')
```

```python
import pandas as pd

# Assuming 'final_df1' is your DataFrame loaded from the CSV file
# For demonstration, let's pretend 'final_df1' is already defined with the necessary data

# 1. Mean and median PM2.5 reading over all sensors
mean_pm25 = final_df1['value'].mean()
median_pm25 = final_df1['value'].median()
print(f"Mean PM2.5 reading: {mean_pm25}")
print(f"Median PM2.5 reading: {median_pm25}")

# 2. Standard deviation of PM2.5 readings
std_dev_pm25 = final_df1['value'].std()
print(f"Standard deviation of PM2.5 readings: {std_dev_pm25}")

# 3. Location_id recording the highest PM2.5 and its reading
highest_pm25 = final_df1.loc[final_df1['value'].idxmax(), ['locationId', 'value']]
print(f"Location ID with the highest PM2.5 reading: {highest_pm25['locationId']}")
print(f"Highest PM2.5 reading: {highest_pm25['value']}")

# 4. Ratio of Community Organization to Governmental Organization entity type
entity_types = final_df1['entity'].value_counts()
ratio_community_to_gov = entity_types['Community Organization'] / entity_types['Governmental Organization']
print(f"Ratio of Community Organization to Governmental Organization: {ratio_community_to_gov}")

# 5. Number of unique sensor stations
unique_sensors = final_df1['locationId'].nunique()
print(f"Number of unique sensor stations: {unique_sensors}")

# 6. Station density per km (assuming the area is 7.5 km² as mentioned)
area_km2 = 7.5  # This is an assumption; replace with actual area if known
station_density = unique_sensors / area_km2
print(f"Station density per km²: {station_density}")

# 7. Daily statistics separately for each entity type
# Convert 'local_time' to datetime if it's not already
final_df1['local_time'] = pd.to_datetime(final_df1['local_time'])

# Group by entity and date, then calculate statistics
daily_stats = final_df1.groupby(['entity', final_df1['local_time'].dt.date]).agg(
    mean_pm25=('value', 'mean'),
    median_pm25=('value', 'median'),
    min_pm25=('value', 'min'),
    max_pm25=('value', 'max'),
    std_pm25=('value', 'std'),
    percentile_75=('value', lambda x: x.quantile(0.75))
).reset_index()

# Print daily statistics for each entity type
print("Daily statistics for each entity type:")
print(daily_stats)

```
